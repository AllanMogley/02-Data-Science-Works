```python
import matplotlib.pyplot as plt
```

```python
# The slices will be ordered and plotted counter-clockwise.
labels = 'Pests', 'Pathogens', 'Weeds', 'Drought', 'Others'
sizes = [18, 16, 34, 12, 20]
colors = ['lightskyblue','yellowgreen', 'gold', 'red', 'lightcoral']
explode = (0, 0, 0.1, 0, 0)  # explode a slice if required

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%')
        
#draw a circle at the center of pie to make it look like a donut
centre_circle = plt.Circle((0,0),0.6,color='black', fc='white',linewidth=1.25)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)


# Set aspect ratio to be equal so that pie is drawn as a circle.
plt.axis('equal')
plt.show()  
```
